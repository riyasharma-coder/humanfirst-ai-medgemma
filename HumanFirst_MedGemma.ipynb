```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_name = "microsoft/BioGPT"
```

```python
!pip install -q sacremoses
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
medical_report = """
Hemoglobin: 10.5 g/dL
WBC Count: 11,000 cells/mm3
Platelet Count: 250,000 cells/mm3

Impression: Mild anemia.
"""
```

```python
prompt = f"""
You are a healthcare explanation assistant.

Your task is to help patients understand medical reports in simple language.

Follow this structure strictly:

1. Simple Summary:
Explain the report in easy everyday language.

2. Key Medical Terms Explained:
Explain any medical terms mentioned.

3. What This Means For You:
Give general understanding only. Do NOT diagnose.

4. Questions You Can Ask Your Doctor:
Suggest helpful questions.

Rules:
- Do NOT provide diagnosis
- Do NOT suggest treatment or medicines
- Do NOT predict diseases
- Use calm, respectful and reassuring tone

Medical Report:
{medical_report}
"""

```

```python
inputs = tokenizer(prompt, return_tensors="pt")

output = model.generate(
    **inputs,
    max_new_tokens=350,
    temperature=0.6,
    do_sample=True
)

response = tokenizer.decode(output[0], skip_special_tokens=True)
print(response)
```

```python
app_py_content = """import streamlit as st
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch # Added import for torch

st.set_page_config(page_title="HumanFirst AI", layout="centered")

st.title("HumanFirst AI")
st.subheader("Understand your medical report in simple language")

st.markdown(
    "⚠️ This tool does not provides medical advice. Always consult a healthcare professional."
)

@st.cache_resource
def load_model():
    model_name = "microsoft/BioGPT"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)
    return tokenizer, model

tokenizer, model = load_model()

medical_text = st.text_area(
    "Paste your medical report below:",
    height=250
)

if st.button("Explain My Report"):
    if medical_text.strip() == "":
        st.warning("Please paste a medical report first.")
    else:
        # Define the prompt template as a regular string, to be formatted later
        prompt = '''
You are a healthcare explanation assistant.

Explain the following medical report in simple language for a patient.

Rules:
- Do not diagnose
- Do not suggest treatment or medicines
- Use calm and supportive tone
- Explain medical terms simply

Medical Report:
{medical_text}
'''.format(medical_text=medical_text)

        inputs = tokenizer(prompt, return_tensors="pt")

        # Ensure correct device placement if CUDA is available
        if torch.cuda.is_available():
            model.to("cuda")
            inputs = {k: v.to("cuda") for k, v in inputs.items()}

        output = model.generate(
            **inputs,
            max_new_tokens=350,
            temperature=0.6,
            do_sample=True
        )

        response = tokenizer.decode(output[0], skip_special_tokens=True)

        st.subheader("Explanation:")
        st.markdown(response)
"""

with open("app.py", "w") as f:
    f.write(app_py_content)

print("app.py created successfully!")
```
